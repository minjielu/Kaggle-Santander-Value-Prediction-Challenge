```python
# This function uses leak to predict target
def _get_leak(df, cols, lag=0):
    cols_1 = []
    cols_2 = []
    for i in np.arange(len(cols)):
        cols_1.extend(cols[i][:-lag-2])
        cols_2.extend(cols[i][lag+2:])
    # trash_key = tuple([0]*(len(new_cols)-2-lag))
    d1 = df[cols_1].apply(tuple, axis=1).to_frame().rename(columns={0: 'key'})
    d2 = df[cols_2].apply(tuple, axis=1).to_frame().rename(columns={0: 'key'})
    # d2['nonzero'] = df[cols_2].apply(lambda x: sum(x != 0),axis=1)
    # d2 = d2[d2['nonzero'] > 0]
    d2['pred'] = df[cols[0][lag]]
    d2 = d2[d2.pred != 0] ### to make output consistent with Hasan's function
    d3 = d2[~d2.duplicated(['key'], keep=False)]
    d4 = d1[~d1.duplicated(['key'], keep=False)]
    d5 = d4.merge(d3, how='inner', on='key')
    # d3 = d3[d3['key'] != trash_key]
    
    return d1.merge(d5, how='left', on='key').pred.fillna(0)
```
